  Mode :-

```python
age_groups = ['15-25', '25-35', '35-45', '45-55', '55-65']
f = [350, 375, 120, 80, 90]

# Step 1: Calculate mid-points
x = []  # mid-points
for group in age_groups:
    lower, upper = map(int, group.split('-'))
    mid = (lower + upper) / 2
    x.append(mid)

# Step 2: Calculate Mean
fx = [f[i] * x[i] for i in range(len(f))]
mean = sum(fx) / sum(f)

# Step 3: Display Result
print("Mean =", round(mean, 2))
```

```python
#step 1: caculate cumulative frequency
cf = []
total = 0
for freq in f:
    total += freq
    cf.append(total)

# Step 2: Find total frequency (N) and N/2
N = sum(f)
Nby2 = N / 2

# Step 3: Find median class (the class where N/2 lies)
for i in range(len(cf)):
    if cf[i] >= Nby2:
        median_class_index = i
        break

# Step 4: Extract required values
L = int(age_groups[median_class_index].split('-')[0])                                             # lower limit of median class
F = cf[median_class_index - 1] if median_class_index > 0 else 0                                                           # cumulative freq before median class
f_m = f[median_class_index]                                                                        # frequency of median class
h = 10                                                                                                                    # class width

# Step 5: Apply Median formula
# Median = L + ((N/2 - F) / f) * h
median = L + ((Nby2 - F) / f_m) * h

# Step 6: Display result
print("Median =", round(median, 2))
```

```python
# Step 1: Find the modal class (highest frequency)
modal_index = f.index(max(f))

# Step 2: Extract required values
L = int(age_groups[modal_index].split('-')[0])                                                                    # lower limit of modal class
f1 = f[modal_index]                                                                                                                      # frequency of modal class
f0 = f[modal_index - 1] if modal_index > 0 else 0                                                                 # frequency before modal class
f2 = f[modal_index + 1] if modal_index < len(f) - 1 else 0                                                                               # frequency after modal class
h = 10                                                                                                            # class width

# Step 3: Apply Mode formula
# Mode = L + ((f1 - f0) / (2*f1 - f0 - f2)) * h
mode = L + ((f1 - f0) / (2*f1 - f0 - f2)) * h

# Step 4: Display result
print("Mode =", round(mode, 2))
```
