 # Given data in question:-

Population mean (μ) = 85%

Sample mean (x̄) = 88%

Standard deviation (σ) = 1

Distribution = Gaussian (Normal)

### Hypothesis :-

H₀ (Null Hypothesis): Mean quality = 85%

H₁ (Alternative Hypothesis): Mean quality > 85%

This is a right-tailed Z test.

### Z-Score Formula :- 

              Z = x̄ − μ / σ
              Z = 88 − 85 / 1
              Z = 3

### Decision Rule :-

For 5% significance level

Critical Z value (right-tailed) = 1.645

Since

     Z = 3 > 1.645
## Conclusion :-

The calculated Z value is greater than the critical value.

So, we Reject Null Hypothesis.

#### His claim is statistically valid.
The hypothesis is accepted — his garden has significantly better quality mangoes.



```python
mu = 85
x_bar = 88
sigma = 1

z = (x_bar - mu) / sigma
print("Z Score =", z)

if z > 1.645:
    print("Reject Null Hypothesis - Claim Accepted")
else:
    print("Fail to Reject Null Hypothesis - Claim Not Accepted")
```
