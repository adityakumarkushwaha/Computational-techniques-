# Confusion Matrix :-
| Actual \ Prediction | **Positve** | **Negative** |
| ------------------- | ----------- | ------------ |
| **True**            |     199     |      1       |
| **False**           |       3     |     247      |


## So:-

TP = 199

FN = 1

FP = 3

TN = 247

### Total samples :-
total = TP + TN + FP + FN

      = 199 + 247 + 3 + 1
      = 450

### Accuracy :-
accuracy = (TP + TN) / total

         = (199 + 247) / 450
         = 446/450
         = 0.9911111111111112

### Precision :-
precision = TP / (TP + FP)         

          = 199 / (199 + 3) 
          = 199 / 202
          = 0.9851485148514851

### Recall :-
recall = TP / (TP + FN)
       
       = 199 / (199 + 1)
       = 199 / 200
       = 0.995

### F1 Score :-
f1_score = 2 * (precision * recall) / (precision + recall)

         = 2 * (0.9851485148514851 * 0.995)/(0.9851485148514851 + 0.995)
         = 2 * (0.9802227722772277 /1.980148514851485 )
         = 2 * 0.4950248756218906
         = 0.9900497512437811

```python
# Confusion matrix values
TP = 199
FN = 1
FP = 3
TN = 247

total = TP + TN + FP + FN

accuracy = (TP + TN) / total

precision = TP / (TP + FP)

recall = TP / (TP + FN)

f1_score = 2 * (precision * recall) / (precision + recall)

print("Accuracy =", accuracy)
print("Precision =", precision)
print("Recall =", recall)
print("F1-Score =", f1_score)
```
